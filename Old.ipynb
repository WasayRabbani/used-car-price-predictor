Before all that stuff, lets simply train this model and see the results

```python
# So i need to encoding because without it, i cant even train simple model
```

```python
# Here what i did is that my value of price was $10,300 i made it 10300 and also made it integer

df['price']=df['price'].apply(lambda x:x.replace('$','').replace(',', ''))
df['price']=df['price'].astype('int')
```

```python
# doing what we did with price column, now with mileage

df['milage']=df['milage'].apply(lambda x:x.replace(',', '').replace('mi.',''))
df['milage']=df['milage'].astype('int')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(['price'],axis=1),df.price,test_size=0.2,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
col=ColumnTransformer(
    transformers=[
        ('OHE',OneHotEncoder(sparse_output=False,dtype='int',drop='first',handle_unknown='ignore'),['brand','model','fuel_type','engine','transmission','ext_col','int_col','accident','clean_title'])
    ],remainder='passthrough'
)
X_train_trf=col.fit_transform(X_train)
X_test_trf=col.transform(X_test)
```

Applying Linear Regression model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train_trf,y_train)
y_pred=lr.predict(X_test_trf)
```

Checking accuracy 

```python
from sklearn.metrics import r2_score
score=r2_score(y_pred,y_test)
print(score)
```
